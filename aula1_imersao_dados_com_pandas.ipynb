```python
import numpy as np

#Criacao de um dataframe de testes
df_salarios = pd.DataFrame({
  "nome": ["Ana", "Bruno", "Carlos", "Daniele","val"],
  "salario": [4000,np.nan, 5000, np.nan, 100000 ]
  })

#calcula e substitui os nulos pela média e arredonda os valores
df_salarios["salario_media"] = df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2))

'''comentatio calcula mediana e substitui os nulos pela mediana
'''
df_salarios["salario_mediana"] = df_salarios["salario"].fillna(df_salarios["salario"].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia":["Segunda ", "Terça", "Quarta,", "Quinta","Sexta" ],
    "Temperatura": [30, np.nan, np.nan,28,27]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
#ffill (fordfill) preenche os dados com o valor do dia anterior
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia":["Segunda ", "Terça", "Quarta,", "Quinta","Sexta" ],
    "Temperatura": [30, np.nan, np.nan,28,27]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()
#bfill preenche os dados com o valor do dia posterior
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Daniele","val"],
    "cidade": ["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]

})
df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Não informado")
df_cidades
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x="senioridade", y="usd")
plt.title("Salario medio por nivel de senioridade")
plt.xlabel("Senioridade")
plt.ylabel("Salario medio anual(USD)")
plt.show()
```

```python
import plotly.express as px
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names="tipo_trabalho",
             values="quantidade",
             title="Proporção dos tipos de trabalho"
)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

# Map numerical values to descriptive labels
remoto_map = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
remoto_contagem['tipo_trabalho'] = remoto_contagem['tipo_trabalho'].replace(remoto_map)

fig = px.pie(remoto_contagem,
             names="tipo_trabalho",
             values="quantidade",
             title="Proporção dos tipos de trabalho"
)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

remoto_map = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
remoto_contagem['tipo_trabalho'] = remoto_contagem['tipo_trabalho'].replace(remoto_map)

fig = px.pie(remoto_contagem,
             names="tipo_trabalho",
             values="quantidade",
             title="Proporção dos tipos de trabalho",
             hole=0.5
)
fig.update_traces(textinfo = "percent+label")
fig.show()
```

```python
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']
salario_por_pais_ds = df_data_scientist.groupby('empresa')['usd'].mean().reset_index()

fig = px.bar(salario_por_pais_ds,
             x='empresa',
             y='usd',
             title='Média Salarial para Cientistas de Dados por País')
fig.update_layout(xaxis_title='País', yaxis_title='Média Salarial Anual (USD)')
fig.show()
```
